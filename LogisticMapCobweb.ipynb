```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

plt.rcParams.update({'axes.labelsize': 18, 'xtick.labelsize': 16, 'ytick.labelsize': 16})

def logistic_map(r, x):
    return r * x * (1 - x) # Function

def cobweb_plot(f, x0, N=50, a=0, b=1): # Cobweb Plot
    ax = plt.subplot(1, 2, 2)
    t = np.linspace(a, b, 1000)
    ax.plot(t, f(t), 'k')
    ax.plot(t, t, "b")
    x_prev = x0 # Initial starting point
    x = f(x0)
    
    for _ in range(N):
        ax.plot([x_prev, x], [x, x], 'r')
        y = f(x)
        ax.plot([x, x], [x, y], 'r')
        ax.plot(x0, f(x0), 'bo')
        x_prev = x
        x = y

    ax.set_xlabel("x(k)")
    ax.set_ylabel("x(k+1)")
    ax.set_xlim(a, b)
    ax.set_ylim(a, b)
    ax.set_aspect('equal', adjustable='box')
    ax.grid(True)

def trajectory_plot(r, x0, N=50, eps=0.5): # Time Series Plot corresponding to Cobweb Plot
    ax = plt.subplot(1, 2, 1)
    eps = 1e-6
    x_vals = [x0]
    x_vals_1 = [x0+eps]
    x1 = x0
    x1_eps = x0+eps
    for _ in range(N):
        x1 = logistic_map(r, x1)
        x1_eps = logistic_map(r, x1_eps)
        x_vals.append(x1)
        x_vals_1.append(x1_eps)
    
    n_vals = np.arange(N + 1)
    ax.plot(n_vals, x_vals_1, 'r-')
    ax.plot(n_vals, x_vals, 'b-')
    ax.set_title(f"r = {r}")
    ax.set_xlabel("k")
    ax.set_ylabel("x(k)")
    ax.set_ylim(0, 1)
    ax.grid(True)
    
def update_combined_plot(r=2.5, x0=0.1): # Combined update function
    plt.clf()
    plt.gcf().set_size_inches(12, 6)
    f = lambda x: logistic_map(r, x)
    trajectory_plot(r, x0)
    cobweb_plot(f, x0)
    plt.tight_layout()
    plt.show()

r_slider = widgets.FloatSlider(min=0, max=4.0, step=0.01, value=2.5, description='r') # Customizable values for r and x0
x0_slider = widgets.FloatSlider(min=0, max=1, step=0.01, value=0.1, description='x0')

r_driver = widgets.IntSlider(min=0, max=800, step=1, value=250)
x0_driver = widgets.IntSlider(min=0, max=200, step=1, value=10)

def update_r(change): r_slider.value = change['new'] / 200 # Fine-tuning
def update_x0(change): x0_slider.value = change['new'] / 200

r_driver.observe(update_r, names='value')
x0_driver.observe(update_x0, names='value')

r_play = widgets.Play(interval=50, value=250, min=0, max=800, step=1)
x0_play = widgets.Play(interval=50, value=10, min=0, max=200, step=1)

widgets.jslink((r_play, 'value'), (r_driver, 'value')) # Play button to allow real-time live update of either r or x0
widgets.jslink((x0_play, 'value'), (x0_driver, 'value'))

ui = widgets.VBox([
    widgets.HBox([r_play, r_slider]),
    widgets.HBox([x0_play, x0_slider])])

out = widgets.interactive_output(update_combined_plot, {'r': r_slider, 'x0': x0_slider})
display(ui, out)
```
