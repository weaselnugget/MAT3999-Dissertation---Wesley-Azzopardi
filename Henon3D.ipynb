```python
import numpy as np
import plotly.graph_objects as go

b_values = np.linspace(-0.5, 0.5, 800)
def Henon_Map(a): # Fixing a, Varying b
    xs, ys, b_s = [], [], []
    for b in b_values:
        x, y = 0.1, 0.1
        for _ in range(1000): # Burn-in Period
            x, y = 1 - a*x*x + y, b*x
            if abs(x) > 10 or abs(y) > 10:
                break # For smoother rendering

        for _ in range(100): # Collect points
            x, y = 1 - a*x*x + y, b*x
            if abs(x) > 10 or abs(y) > 10:
                break
                
            xs.append(x)
            ys.append(y)
            b_s.append(b)
    return xs, ys, b_s

a_values = np.linspace(0, 2, 41)  # Step = 0.05
fig = go.Figure()
for i, a in enumerate(a_values): # Create traces for each a value
    xs, ys, a_s = Henon_Map(a)
    fig.add_trace(
        go.Scatter3d(
            x=xs,
            y=ys,
            z=a_s,
            mode='markers',
            marker=dict(size=1, color='#EB5E00'),
            visible=(i == 0),
            name=f"a = {a:.2f}" ))
steps = []

for i, a in enumerate(a_values): # Customizable values for a
    steps.append(
        dict(
            method="update",
            args=[{"visible": [j == i for j in range(len(a_values))]}],
            label=f"{a:.2f}"),)

fig.update_layout(
    sliders=[dict(
        active=0,
        currentvalue={"prefix": "a = "},
        steps=steps)],
    scene=dict(
        xaxis_title=r'x_n',
        yaxis_title=r'y_n',
        zaxis_title=r'b',),
    margin=dict(l=0, r=0, b=0, t=40),)
fig.show()
fig.write_html("Henon3DBifurcationPlot_Vary_a.html", include_plotlyjs='cdn', full_html=True) # Save as an Open File
```
