```python
import numpy as np
import cmath
from math import sqrt
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider

# ---- For Henon Map ----

def fixed_points(a,b):
    A, B, C = a, (1-b), -1
    disc = B**2 - 4*A*C
    if disc < 0: return []
    x1 = (-B + sqrt(disc))/(2*A)
    x2 = (-B - sqrt(disc))/(2*A) # Solve by Quadratic Formula
    return [(x1, b*x1), (x2, b*x2)] # Fixed Points

def eigen(a,b,x):
    term = (a*x)**2 + b
    r = cmath.sqrt(term)
    return -a*x + r, -a*x - r # lam1 and lam2, written this way to prevent error in interaction

def interactive_sliders(a=1.4,b=0.3): # Eig sliders
    vals_real, vals_imag = [], []
    fps = fixed_points(a,b)
    for x,y in fps:
        l1, l2 = eigen(a,b,x)
        vals_real += [l1.real, l2.real]
        vals_imag += [l1.imag, l2.imag]

    angles = np.linspace(0, 2*np.pi, 600) # Unit Circle
    xc = np.cos(angles)
    yc = np.sin(angles)

    fig, ax = plt.subplots()
    ax.plot(xc, yc, linewidth=2)
    ax.scatter(vals_real, vals_imag, s=40, color='k')
    ax.axhline(0, linewidth=0.5)
    ax.axvline(0, linewidth=0.5)
    ax.set_aspect('equal', adjustable='box')
    ax.set_xlim(-4,4)
    ax.set_ylim(-4,4)
    plt.show()

interact(
    interactive_sliders,
    a=FloatSlider(value=1.4, min=0, max=2, step=0.01, description='a'),
    b=FloatSlider(value=0.3, min=-0.5, max=0.5, step=0.01, description='b'))
```

```python
import numpy as np
import cmath
from math import sqrt
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider

# ---- For ZS ----
def fixed_points_ZS(a, b):
    fps = [(0, 0)]
    if a < -1: 
        y = np.sqrt(-a-1)
        fps.append(((1-b)*y, y))
        fps.append((-(1-b)*y, -y))
    return fps

def jacobian_ZS(a, b, x, y):
    denom = 1+y**2
    J = np.array([[-a/denom, 2*a*x*y/(denom**2)], [1, b]])
    return J

def eigen_ZS(a, b, x, y):
    J = jacobian_ZS(a, b, x, y)
    return np.linalg.eigvals(J)

def interactive_ZS(a=3.8, b=0.6): # Eig Sliders 
    vals_real, vals_imag = [], []

    fps = fixed_points_ZS(a, b)
    for x, y in fps:
        eigs = eigen_ZS(a, b, x, y)
        for l in eigs:
            vals_real.append(l.real)
            vals_imag.append(l.imag)

    theta = np.linspace(0, 2*np.pi, 600) # Unit Circle
    xc, yc = np.cos(theta), np.sin(theta)

    fig, ax = plt.subplots()
    ax.plot(xc, yc, linewidth=2)
    ax.scatter(vals_real, vals_imag, s=40, color='k')

    ax.axhline(0, linewidth=0.5)
    ax.axvline(0, linewidth=0.5)

    ax.set_aspect('equal', adjustable='box')
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)

    ax.set_title("ZS map eigenvalues vs unit circle")
    plt.show()

interact(
    interactive_ZS,
    a=FloatSlider(value=3.8, min=-6, max=6, step=0.01, description='a'),
    b=FloatSlider(value=0.6, min=-1.2, max=1.2, step=0.01, description='b'))
```
