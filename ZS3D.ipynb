```python
import numpy as np
import plotly.graph_objects as go

b_values = np.linspace(-1.2, 1.2, 800)
def ZS_Map(a): # Fixing a, Varying b
    xs, ys, b_s = [], [], []
    for b in b_values:
        x, y = 0.1, 0.1 # Burn-in Period
            x, y = -(a*x)/(1+y**2), x+b*y
            if abs(x) > 10 or abs(y) > 10:
                break # For smoother rendering
                
        for _ in range(100): # Collect Points
            x, y = -(a*x)/(1+y**2), x+b*y
            if abs(x) > 10 or abs(y) > 10:
                break
                
            xs.append(x)
            ys.append(y)
            b_s.append(b)
    return xs, ys, b_s

a_values = np.linspace(-6, 6, 121)  # Step = 0.01 (Loss of Memory Space on my PC otherwise)
fig = go.Figure()
for i, a in enumerate(a_values): # Create traces for each a value
    xs, ys, a_s = ZS_Map(a)
    fig.add_trace(
        go.Scatter3d(
            x=xs,
            y=ys,
            z=a_s,
            mode='markers',
            marker=dict(size=1, color='#EB5E00'),
            visible=(i == 0),
            name=f"a = {a:.2f}" ))
steps = []

for i, a in enumerate(a_values): # Customizable values for a
    steps.append(
        dict(
            method="update",
            args=[{"visible": [j == i for j in range(len(a_values))]}],
            label=f"{a:.2f}"),)

fig.update_layout(
    sliders=[dict(
        active=0,
        currentvalue={"prefix": "a = "},
        steps=steps)],
    scene=dict(
        xaxis_title=r'x_n',
        yaxis_title=r'y_n',
        zaxis_title=r'b',),
    margin=dict(l=0, r=0, b=0, t=40),)
fig.show()
fig.write_html("ZS3DBifurcationPlot_Vary_a.html", include_plotlyjs='cdn', full_html=True) # Save as an Open File
```

```python
import numpy as np
import plotly.graph_objects as go

a_values = np.linspace(-6, 6, 800)
def ZS_Map(b): # Fixing a, Varying b
    xs, ys, a_s = [], [], []
    for a in a_values:
        x, y = 0.1, 0.1
        for _ in range(500): # Burn-in Period
            x, y = -(a*x)/(1+y**2), x+b*y
            if abs(x) > 10 or abs(y) > 10:
                break 
                
        for _ in range(100): # Collect Points
            x, y = -(a*x)/(1+y**2), x+b*y
            if abs(x) > 10 or abs(y) > 10:
                break
                
            xs.append(x)
            ys.append(y)
            a_s.append(a)
    return xs, ys, a_s

b_values = np.linspace(-1.2, 1.2, 49)  # Step = 0.05
fig = go.Figure()
for i, b in enumerate(b_values): # Create traces for each b value
    xs, ys, b_s = ZS_Map(b)

    fig.add_trace(
        go.Scatter3d(
            x=xs,
            y=ys,
            z=b_s,
            mode='markers',
            marker=dict(size=1, color='#EB5E00'),
            visible=(i == 0),
            name=f"b = {b:.2f}" ))
steps = []

for i, b in enumerate(b_values):
    steps.append(
        dict(
            method="update",
            args=[{"visible": [j == i for j in range(len(b_values))]}],
            label=f"{b:.2f}"),)

fig.update_layout(
    xaxis_range=[-10,10],
    yaxis_range=[-10,10],
    sliders=[dict(
        active=0,
        currentvalue={"prefix": "b = "},
        steps=steps)],
    scene=dict(
        xaxis=dict(title=r'x_n'),
        yaxis=dict(title=r'y_n'),
        zaxis=dict(title=r'a')),
    margin=dict(l=0, r=0, b=0, t=40),)
fig.show()
fig.write_html("ZS3DBifurcationPlot_Vary_b.html", include_plotlyjs='cdn', full_html=True) # Save as an Open File
```
