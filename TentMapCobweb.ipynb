```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

plt.rcParams.update({'axes.labelsize': 18, 'xtick.labelsize': 16, 'ytick.labelsize': 16})
def tent_map(a, x): # Function
    x = np.asarray(x)  # Ensure scalar
    return np.where(x < 0.5, a*x, a*(1-x))

def cobweb_plot(f, x0, N=50, a=0, b=1): # Cobweb Plot
    ax = plt.subplot(1, 1, 1)
    t = np.linspace(a, b, 1000)
    ax.plot(t, f(t), 'k')
    ax.plot(t, t, "b")

    x_prev = x0 # Initial starting point
    x = f(x0)
    
    for _ in range(N):
        ax.plot([x_prev, x], [x, x], 'r')
        y = f(x)
        ax.plot([x, x], [x, y], 'r')
        ax.plot(x0, f(x0), 'bo')
        x_prev = x
        x = y
        
    ax.set_xlabel("x(k)")
    ax.set_ylabel("x(k+1)")
    ax.set_xlim(a, b)
    ax.set_ylim(a, b)
    ax.set_aspect('equal', adjustable='box')
    ax.grid(True)

def update_plot(a=1.5, x0=0.1): # Update Function
    plt.clf()
    plt.gcf().set_size_inches(12, 6)
    f = lambda x: tent_map(a, x)
    cobweb_plot(f, x0)
    plt.tight_layout()
    plt.show()

a_slider = widgets.FloatSlider(min=0.0, max=2.0, step=0.01, value=0.5, description='a') # Customizable values for r and x0
x0_slider = widgets.FloatSlider(min=0.0, max=1.0, step=0.01, value=0.1, description='x₀')

a_driver = widgets.IntSlider(min=0, max=400, step=1, value=50)
x0_driver = widgets.IntSlider(min=0, max=200, step=1, value=10)

def update_a(change): a_slider.value = change['new'] / 200 # Fine-tuning
def update_x0(change): x0_slider.value = change['new'] / 200

a_driver.observe(update_a, names='value')
x0_driver.observe(update_x0, names='value')

a_play = widgets.Play(interval=50, value=50, min=0, max=400, step=1)
x0_play = widgets.Play(interval=50, value=10, min=0, max=200, step=1)

widgets.jslink((a_play, 'value'), (a_driver, 'value'))
widgets.jslink((x0_play, 'value'), (x0_driver, 'value')) # Play button to allow real-time live update of either a or x0

ui = widgets.VBox([
    widgets.HBox([a_play, a_slider]),
    widgets.HBox([x0_play, x0_slider])])

out = widgets.interactive_output(update_plot, {'a': a_slider, 'x0': x0_slider})
display(ui, out)
```
