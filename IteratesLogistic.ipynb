```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

plt.rcParams.update({'axes.labelsize': 18, 'xtick.labelsize': 16, 'ytick.labelsize': 16})

def logistic(r, x): # Function
    return r * x * (1 - x)

def logistic_prime_map(r, x): # 1st Derivative
    return r * (1 - 2 * x)
    
def logistic_map(r, x, n): # nth Iterate
    for _ in range(n):
        x = logistic(r, x)
    return x

def fixed_points(r):
    pts = [0, 1 - 1/r]  # Fixed states of f(x)
    if r > 3:  # Period-2
        disc = (r + 1) * (r - 3)
        if disc >= 0:
            sqrt_disc = np.sqrt(disc)
            pts.extend([
                (r + 1 + sqrt_disc) / (2 * r),
                (r + 1 - sqrt_disc) / (2 * r)])
    return np.array(pts)

def logistic_plot(f, ff, f4, r, points, a=0, b=1):
    plt.clf()
    fig = plt.gcf()
    fig.set_size_inches(6, 6)
    
    t = np.linspace(a, b, 1000)
    plt.plot(t, f(t), 'k--', label='f(x)')
    plt.plot(t, t, 'b', label='y = x')
    
    if r <= 3: # If true, display only the first iterate
        x_star = points[1]
        slope = logistic_prime_map(r, x_star)
        plt.plot(t, ff(t), 'k', label='f(f(x))')
        plt.plot(points[1], points[1], 'bo', markersize=15)
        plt.plot(points[0], points[0], 'ro', markersize=10)

    if r > 3: # If true, add the second iterate to display
        plt.plot(t, ff(t), 'k', label='f(f(x))')
        plt.plot(points[1], points[1], 'ro', markersize=15)
        plt.plot(points[2:4], points[2:4], 'bo', markersize=15)

    if r > 1 + np.sqrt(6): # If true, add the third iterate to display
        plt.plot(t, f4(t), 'b--', label='f4(x)')

    plt.xlabel("x(k)")
    plt.ylabel("x(k+2)")
    plt.xlim(a, b)
    plt.ylim(a, b)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True)

def update_plot(r=3):
    f = lambda x: logistic_map(r, x, 1)
    ff = lambda x: logistic_map(r, x, 2)
    f4 = lambda x: logistic_map(r, x, 4)
    points = fixed_points(r)
    logistic_plot(f, ff, f4, r, points)
    plt.tight_layout()
    plt.show()

r_slider = widgets.FloatSlider(min=2.5, max=4, step=0.01, value=3, description='r') # Widgets
r_play = widgets.Play(interval=50, value=250, min=0, max=800, step=1)
r_driver = widgets.IntSlider(min=0, max=800, step=1, value=250)

def update_r(change):
    r_slider.value = change['new'] / 200 # Fine-tuning

r_driver.observe(update_r, names='value')
widgets.jslink((r_play, 'value'), (r_driver, 'value')) # Play button to allow real-time live update

ui = widgets.VBox([
    widgets.HBox([r_play, r_slider])])
out = widgets.interactive_output(update_plot, {'r': r_slider})
display(ui, out)
```
